```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report,r2_score,confusion_matrix
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('/content/Real estate.csv')
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
sns.scatterplot(x='X2 house age',y='Y house price of unit area',data=df)
```

```python
sns.histplot(df['Y house price of unit area'],bins=30,color='red')
```

```python
x=df.drop(['Y house price of unit area'],axis=1)
y=df['Y house price of unit area']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
reg_pred=reg.predict(x_test)
print('r2',r2_score(y_test,reg_pred))
```

```python
scale=StandardScaler()
x_scaled=scale.fit_transform(x)
kmeans=KMeans(n_clusters=3,random_state=42)
df['segment']=kmeans.fit_predict(x_scaled)
```

```python
pca=PCA(n_components=2)
reduc=pca.fit_transform(x_scaled)
plt.figure(figsize=(10,8))
sns.scatterplot(x=reduc[:,0],y=reduc[:,1], hue=df['segment'], palette='Set2')
plt.title('costumer cluster')
```

```python
df['interested'] = np.where(df['Y house price of unit area'] < df['Y house price of unit area'].median(), 1, 0)
df['interested']
```

```python
x=x
y=df['interested']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=42)
model1=RandomForestClassifier()
model1.fit(xtrain,ytrain)
pred1=model1.predict(xtest)
print(classification_report(ytest,pred1))
sns.heatmap(confusion_matrix(ytest,pred1),annot=True,fmt='d')
plt.plot(cmap='Blues')
plt.title("مصفوفة الالتباس لتصنيف العملاء")
plt.grid(False)
plt.show()
```
